```python
import qiskit.tools.jupyter
%qiskit_version_table
%qiskit_copyright

import numpy as np
from qiskit import *
from qiskit.providers.aer import AerSimulator
from qiskit.visualization import plot_histogram

simulator = AerSimulator()
```

```python
circuit = QuantumCircuit(8,2)
circuit.x(0)
circuit.x(1)
circuit.x(2)
circuit.barrier()
circuit.cx(0,3)
circuit.cx(1,3)
circuit.ccx(0,1,4)
circuit.barrier()
circuit.cx(2,7)
circuit.cx(3,7)
circuit.ccx(2,3,5)
circuit.barrier()
circuit.cx(4,6)
circuit.cx(5,6)
circuit.ccx(4,5,6)
circuit.barrier()
circuit.measure(7,0)
circuit.measure(6,1)

result = execute(circuit,simulator,shots=1024).result()
counts = result.get_counts(circuit)
print(f"\nTotal counts for 0 and 1 are:{counts}\n")
plot_histogram(counts)
```
