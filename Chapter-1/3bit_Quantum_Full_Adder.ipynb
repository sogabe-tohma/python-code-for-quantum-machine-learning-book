```python
import qiskit.tools.jupyter
%qiskit_version_table
%qiskit_copyright

import numpy as np
from qiskit import *
from qiskit.providers.aer import AerSimulator
from qiskit.visualization import plot_histogram

simulator = AerSimulator()
```

```python
circuit = QuantumCircuit(11,3)
circuit.x(0)
circuit.x(1)
circuit.x(4)
circuit.x(5)
circuit.barrier()
circuit.cx(0,2)
circuit.cx(1,2)
circuit.ccx(0,1,3)
circuit.barrier()
circuit.cx(4,6)
circuit.cx(5,6)
circuit.ccx(4,5,7)
circuit.barrier()
circuit.cx(3,8)
circuit.cx(6,8)
circuit.ccx(3,6,9)
circuit.barrier()
circuit.cx(7,10)
circuit.cx(9,10)
circuit.ccx(7,9,10)
circuit.barrier()
circuit.measure(10,2)
circuit.measure(8,1)
circuit.measure(2,0)

result = execute(circuit,simulator,shots=1024).result()
counts = result.get_counts(circuit)
print(f"\nTotal counts for 0 and 1 are:{counts}\n")
plot_histogram(counts)
```
