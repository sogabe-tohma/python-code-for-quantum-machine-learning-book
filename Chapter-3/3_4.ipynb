```python
# 3.4 Distance by SWAP test
import numpy as np
from qiskit import *
%matplotlib inline
from qiskit.visualization import plot_histogram
from math import pi

# SWAP test for distance
# all vectors should be in the first quadrant

nshots=10000 # number of shots

# create the full circuit
a=np.array([1.,0.])
b=np.array([1./np.sqrt(2.),1./np.sqrt(2.)])

qr=QuantumRegister(3,'q')
cr=ClassicalRegister(1,'c')

circuit = QuantumCircuit(qr,cr)
circuit.initialize(a,[1])
circuit.initialize(b,[2])
circuit.h(0)
circuit.cswap(qr[0], qr[1], qr[2])
circuit.h(0)

circuit.measure(range(1),range(1))
circuit.draw('mpl')

```

```python

# result 
backend = Aer.get_backend('statevector_simulator')
job = execute(circuit, backend, shots=nshots)
result = job.result()
counts = result.get_counts(circuit)
plot_histogram(counts)


```

```python
# inner product
ab2=(counts['0']-counts['1'])/nshots

# distance
print("Dist(a,b)**2=",2*(1-np.sqrt(ab2)))
print("Distance(a,b)**2 by Numpy",(np.linalg.norm(a-b))**2)
```
